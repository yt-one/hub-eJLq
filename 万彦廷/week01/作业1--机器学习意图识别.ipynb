```python
import pandas as pd
import jieba
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

## 机器学习路线

```python
raw_data = pd.read_csv("dataset.csv", header=None, delimiter="\t", names=["content", "type"])
raw_data.head()
```

```python
raw_data["content"] = raw_data["content"].apply(lambda text: " ".join(jieba.lcut(text)))
```

```python
text_preprocess = CountVectorizer()

feature_vectors = text_preprocess.fit_transform(raw_data["content"].values)
feature_vectors.shape
```

```python
model_lr = LogisticRegression()
model_rf = RandomForestClassifier(max_leaf_nodes=64, n_estimators=200, max_features="sqrt")
model_knn = KNeighborsClassifier()

models = [model_lr, model_rf, model_knn]
```

```python
for model in models:
    model.fit(feature_vectors, raw_data["type"])
```

```python
for model in models:
    print(model.score(feature_vectors, raw_data["type"]))
```

```python
# 预测
test_text = "帮我放一首陶喆的普通朋友"

for model in models:
    prediction = model.predict(text_preprocess.transform([" ".join(jieba.lcut(test_text))]))
    print(prediction)
```
